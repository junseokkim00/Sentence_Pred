```python
in_file_path = '/kaggle/input/netorike/'

out_file_path = '/kaggle/working/'
```

```python
def unlabeling(file_name):
    dataline = pd.DataFrame(columns = ['label', 'txt'])
    
    f = open(file_name, 'r', encoding = 'euc_kr')
    while True:
        line = f.readline()
        if not line: break

        line = line.strip()

        res = line.split(sep='.' ,maxsplit=1)
        if (len(res) == 2) and (res[0][0] != '['):
            df_line = pd.DataFrame([res], columns = ['label', 'txt'])
            dataline = pd.concat([dataline, df_line], ignore_index = True)

    f.close()
    
    return dataline
```

```python
unlabeling(in_file_path+'10.txt')
```

```python
# netorike directory 내부의 txt파일들을 [label, 문장] 형식의 dataframe형식으로 처리

datas = pd.DataFrame(columns = ['label', 'txt'])
    
for dirname, _, filenames in os.walk(in_file_path):
    # - sort the order of files which are in the directory.
    #   - not dictionary order, but numbering order
    #      - (ex. 10, 20, 30, 100, 180, ... // not like 10, 100, 180, 20, 30, ...)
    
    file_tuple = zip( list(map(int,[sen[:-4] for sen in filenames])), filenames ) 
    for num, filename in sorted(file_tuple, key = lambda x : x[0]):
        print(filename)
        file = os.path.join(dirname, filename)
        
        # 각 파일 unlabling 함수에 넣고 datas에 합쳐줌
        datas = pd.concat([datas, unlabeling(file)], ignore_index = True) 
```
