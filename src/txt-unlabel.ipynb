```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
import chardet
# for dirname, _, filenames in os.walk('/kaggle/input'):
#     for filename in filenames:
#         print(os.path.join(dirname, filename))
```

```python
input_file_path = 'C:/Users/____fill____'
output_file_path = 'C:/Users/____fill____'
```

```python
# preprocessing data;       - leave SENTENCES and get rid of any other things, because ONLY SENTENCES should be used training data.
#   ex) "[1. A B then B C] 1. I am a boy. Then many boys are alike me."  -->  "I am a boy. Then many boys are alike me."

def unlabeling(file_name):
    dataline = pd.DataFrame(columns = ['label', 'txt'])
    
    # open file for checking the <.txt> file's encoding type (utf8, utf16, ...)
    rawdata = open(file_name, 'rb').read()
    result = chardet.detect(rawdata)
    enc = result['encoding']
    
    # preprocessing
    f = open(file_name, 'r', encoding = enc)
    while True:
        line = f.readline()
        if not line: break

        line = line.strip()

        res = line.split(sep='.' ,maxsplit=1)
        if (len(res) == 2) and (res[0][0] != '['):
            df_line = pd.DataFrame([res], columns = ['label', 'txt'])
            dataline = pd.concat([dataline, df_line], ignore_index = True)

    f.close()
    
    return dataline
```

```python
# make a dataframe with [label, sentence] form, from <.txt> files which are inside of input directory

datas = pd.DataFrame(columns = ['label', 'txt'])
    
for dirname, _, filenames in os.walk(input_file_path):
    
    # - sort the order of files which are in the directory.
    for filename in sorted(filenames):
        
        if filename == 'output_datas.txt' :
            continue
        
        print('processing ' + filename + '..')
        file = os.path.join(dirname, filename)
        
        # 각 파일 unlabling 함수에 넣고 datas에 합쳐줌
        datas = pd.concat([datas, unlabeling(file)], ignore_index = True) 
```
