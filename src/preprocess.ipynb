```python
import pandas as pd
```

```python
from collections import defaultdict
from transformers import PreTrainedTokenizerFast

tokenizer = PreTrainedTokenizerFast.from_pretrained('skt/kogpt2-base-v2')
newList = defaultdict(list)
newList['prev'] = []
newList['next'] = []
notList = []
allList = []
with open('./data/generated_sents.txt','r') as f:
    for line in f:
        line = line[2:]
        line = line.replace('<','')
        line = line.replace('>','')
        if len(tokenizer.encode(line)) > 50:
            continue
        allList.append(line)
        line = line.split()
        print(line)
        if '그런데' not in line:
            notList.append(line)
            continue
        idx = line.index('그런데')
        prev, _next = line[:idx+1], line[idx+1:]
        newList['prev'].append(' '.join(prev))
        newList['next'].append(' '.join(_next))
allList
```

```python


max_length=0
for line in allList:
    length = len(tokenizer.encode(line))
    if max_length < length:
        max_length = length
max_length
```

```python
df = pd.DataFrame(newList)
df
```

```python
df2 = pd.read_csv('./data/train2.csv')
df2
```

```python
df2 = df2[['prev','next']]
df2
```

```python
list_ = df2['prev'] +' '+df2['next']
list_
```

```python
idxExclude = []
for idx in range(len(list_)):
    if len(tokenizer.encode(list_[idx])) > 50:
        idxExclude.append(idx)
print(len(idxExclude))
```

```python
df2 = df2.drop(idxExclude)
df2 = df2.reset_index(drop=True)
df2
```

```python
df2 = pd.concat([df,df2])
df2 = df2.reset_index(drop=True)
df2.to_csv('./data/final_train.csv')
```
