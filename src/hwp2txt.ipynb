```python
# path of directory of <.hwp> files, which you want to convert to <.txt>.
hwp_dir_path = "C:/Users/dcba6/Desktop/창업/nlp/data/1/1/"

# path of directory where you want to store <.txt> files.
txt_dir_path = 'C:/Users/dcba6/Desktop/창업/nlp/data/txt/1/'
```

```python
# convert hwp to txt
def hwp2txt(file_path, file_name, txt_dir_path = txt_dir_path):
    
    # os.path.splitext() -> [0] : filename // [1] : type of file, such as .txt, .hwp, .pdf, ...
    filename_without_hwp = os.path.splitext(file_name)[0]
    
    # whole file path. ( ex. C:\Users\dcba\Desktop\abcd.hwp )
    file = file_path+file_name
    
    # file path & name, which will be generated. ( ex. C:\Users\dcba\Desktop\txt_dir\abcd.txt )
    gen_file = txt_dir_path+filename_without_hwp+".txt"
    
    # converting command
    terminal_command = f"hwp5txt --output \"{gen_file}\" \"{file}\""
    os.system(terminal_command)
    
    # return file path
    return gen_file
```

```python
for dirname, _, filenames in os.walk(hwp_dir_path):
    
    for filename in sorted(filenames):
        print('converting ' + '\"' + filename + '\" ..' )
        file = os.path.join(dirname, filename)
        
        # if the type of file is '.hwp', call "hwp2txt" function
        if os.path.splitext(file)[1] == '.hwp':
            hwp2txt(dirname, filename)
```
