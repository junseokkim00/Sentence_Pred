```python
import os
import pandas as pd
```

```python
file_path = './그러나/'
arr = os.listdir('./그러나/')
```

```python
from collections import defaultdict
prompts = defaultdict(list)
for filename in arr:
    with open(file_path+filename, 'r') as f:
        prompts[filename] = f.readlines()
```

```python
for filename in prompts:
    for i in range(len(prompts[filename])):
        string = prompts[filename][i].strip('\n')
        string = string.split()
        idx = string.index('그러나')
        prompts[filename][i] = '<'+' '.join(string[:idx]) + '>' +' 그러나 '+'<'+' '.join(string[idx+1:])+'>'
```
